```python
import pandas as pd
import cv2
import matplotlib.pyplot as plt
import json
import numpy as np
```

## Reading Image Using OpenCV

```python
#read an image using opencv - Note opencv uses BGR instead of RGB - need to take this into account as our images are originally in RGB

img = cv2.imread('images/42.6804_2.9496714.png')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
with open('metadata.json') as f:
    meta = json.load(f)
```

```python
meta['42.6804_2.9496714.png'].keys()
```

```python
meta['42.6804_2.9496714.png']['bounds_x_y']
```

```python
def parse_points(lod):
    points = []
    for point in lod:
        coordinate = []
        coordinate.append(round(point['x'],0))
        coordinate.append(round(point['y'],0))
        points.append(coordinate)
    return points
```

```python
pts = np.array(parse_points(meta['42.6804_2.9496714.png']['bounds_x_y']))
pts = pts.reshape((-1, 1, 2))
```

```python
image = cv2.polylines(img_rgb,np.int32([pts]),True,(0,0,255),2)
```

```python
plt.imshow(image)
```
