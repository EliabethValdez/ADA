```python
import pandas as pd
```

```python
month1 = pd.read_csv('data/train_month_1.csv')
month2 = pd.read_csv('data/train_month_2.csv')
month3 = pd.read_csv('data/train_month_3_with_target.csv')
```

```python
month1.head()
```

```python
month2.head()
```

```python
month3.head()
```

```python
notequal = []
for col in list(month1):
    notequal.append(not month1[col].equals(month2[col]))
notequal[0] = True
print(notequal)
```

```python
notequal2 = []
for col in list(month1):
    notequal2.append(not month1[col].equals(month3[col]))
notequal2.append(True)
notequal2[0] = True

print(notequal2)
```

```python
month3 = month3.add_suffix("_3")
month3 = month3.rename(columns={"client_id_3":"client_id", "target_3":"target"})
month3.head()
```

```python
full_df = pd.merge(month1, month2.loc[:,notequal], left_on="client_id", right_on="client_id", how="inner", suffixes=["_1", "_2"])
full_df = pd.merge(full_df, month3.loc[:,notequal2], left_on="client_id", right_on="client_id", how="inner")
full_df.head()
```

```python
month1.loc[:, notequal].columns
```

```python

#for each column with different months, we want to combine them into a single column with sequential binary values
for col in list(month1.loc[:,notequal].columns):
    if col != "client_id":
        
        full_df[str(col) + "_seq"] = full_df.filter(regex="^"+str(col)+"_[0-9]").apply(lambda x: ''.join(x.astype(str)), axis=1)
        
```

```python
# for each sequential column, create a crosstable
cross_table_lst = []
for col in list(full_df.filter(regex="_seq", axis=1).columns):
    cross_table = pd.crosstab(full_df[col], full_df['target'])
    cross_table_lst.append(cross_table)
```

# Crosstable for First Variable

```python
cross_table_lst[0].index.name
```

# Pearson Chi-Square Test for Independence

For each cross-table in the list, perform a pearson's chi-square test against the target variable to see if they are independent or not

```python
from scipy import stats
```

```python
# for each cross-table, if the chi-square test yields a significant p-value -> the variable and the target is not independent and we want to append to list
sig_lst = []
for ct in cross_table_lst:
    chi_sq, p_val, df, exp= stats.chi2_contingency(cross_table_lst[0])
    if p_val < 0.05:
        sig_lst.append(ct.index.name)
```

```python
print(sig_lst)
```

```python
print(len(sig_lst))
print(len(cross_table_lst))
```
