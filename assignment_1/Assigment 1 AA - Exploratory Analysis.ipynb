```python
dt3['customer_since_all_d']=pd.to_datetime(dt3['customer_since_all'], format='%Y-%m-%d')
dt3['customer_birth_date_d']=pd.to_datetime(dt3['customer_birth_date'], format='%Y-%m-%d')
dt3.sort_values(by=['customer_since_all_d','customer_birth_date_d'], ascending=False)
```

```python
dt3['age']=2018-pd.DatetimeIndex(dt3['customer_birth_date_d']).year
dt3['mob']=((pd.to_datetime('2018-10-01')-dt3['customer_birth_date_d'])/np.timedelta64(1, 'M')).astype('int')
dt3.head()
```

```python
dt3['insurance']=dt3[['has_insurance_21','has_insurance_23','has_life_insurance_fixed_cap','has_life_insurance_decreasing_cap','has_fire_car_other_insurance']].apply(lambda row: row.max(),axis=1)
dt3['savings']=dt3[['has_pension_saving','has_savings_account','has_savings_account_starter','has_current_account_starter']].apply(lambda row: row.max(),axis=1)
dt3['loan']=dt3[['has_personal_loan','has_mortgage_loan','has_current_account']].apply(lambda row: row.max(),axis=1)
dt3['has_account']=dt3[['insurance','savings','loan']].apply(lambda row: row.max(),axis=1)
dt3['bal_insurance']=dt3[['bal_insurance_21','bal_insurance_23']].apply(lambda row: row.sum(),axis=1)
dt3['bal_savings']=dt3[['bal_pension_saving','bal_savings_account','bal_savings_account_starter','bal_current_account_starter']].apply(lambda row: row.sum(),axis=1)
dt3['bal_loan']=dt3[['bal_personal_loan','bal_mortgage_loan','bal_current_account']].apply(lambda row: row.sum(),axis=1)
dt3['debt_ratio']=np.where(dt3['bal_loan']/dt3['bal_savings']>=0,dt3['bal_loan']/dt3['bal_savings'],0)
dt3['gender'] = np.where(dt3['customer_gender']==1, 1, 0)
dt3[['client_id','target','insurance','savings','loan','has_account','bal_insurance','bal_savings','bal_loan','debt_ratio','age','mob','customer_relationship','customer_self_employed','customer_occupation_code','gender']]
```

```python
y=dt3['target']
X=dt3[['insurance','savings','loan','bal_insurance','bal_savings','bal_loan','debt_ratio','homebanking_active']]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4, random_state=42)
lr_model = LogisticRegression(multi_class='multinomial', solver='lbfgs');
lr_model.fit(X_train, y_train)
y_pred_lr = lr_model.predict(X_test)
lr_model.score(X_train, y_train)
```

```python
# Import necessary modules
from sklearn.metrics import roc_curve

# Compute predicted probabilities: y_pred_prob
y_pred_prob = lr_model.predict_proba(X_test)[:,1]

# Generate ROC curve values: fpr, tpr, thresholds
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)

# Plot ROC curve
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.show()
```
